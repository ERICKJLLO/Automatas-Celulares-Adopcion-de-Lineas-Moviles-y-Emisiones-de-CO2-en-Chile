## **Metodologia**
### **Modelo de Autómatas Celulares**
Los autómatas celulares son una técnica de modelado computacional que permite simular sistemas complejos mediante una cuadrícula de celdas, donde cada celda representa una unidad poblacional (como un individuo o una familia) que puede cambiar de estado en función de reglas locales. En este caso, cada celda representa a una persona en Chile que puede adoptar o no una línea móvil.

### Parámetros del Modelo:
* **Tamaño de la cuadrícula (size):** Una matriz de 250x250 celdas que representa la población de Chile.
* Número de años (years): La simulación cubre el período de 1990 a 2023.
* **Probabilidad de adquisición (pa_):** Representa la probabilidad de que una celda adopte una línea móvil en función de los vecinos que ya tienen una.
* Probabilidad de desconexión (pd_): **texto en negrita** Captura la posibilidad de que una celda con línea móvil cancele el servicio, modelando la saturación del mercado.

**Estado Inicial:** La simulación comienza con pocas celdas activas (personas con línea móvil), representando la adopción baja de telefonía en 1990.

### **Análisis de Datos Históricos**
Para validar la simulación, se compara la curva de adopción simulada con datos históricos de suscripciones móviles en Chile. Estos datos se extraen de fuentes confiables y se representan gráficamente. Además de esto:
*  Se utilizó una regresión lineal para estimar la tendencia en el crecimiento de las suscripciones, ajustando una línea de regresión a los datos históricos.
*  Se empleó la biblioteca Seaborn para crear gráficos que visualicen la relación entre el PIB, emisiones de CO₂ y el crecimiento de la telefonía móvil.


## **Modelo de Autómatas Celulares**
El modelo de autómatas celulares simula el crecimiento en la adopción de líneas móviles en Chile, utilizando como punto de partida el número de líneas registradas en 1990. Los parámetros principales del modelo son:
- **Tamaño de la matriz**: Representa el espacio total de simulación (250x250 celdas).
- **Años de simulación**: Se simulan 33 años, de 1990 a 2022.
- **Probabilidad de adquisición (`pa_`)**: La probabilidad de que una celda adquiera una línea móvil, establecida en 0.61.
- **Probabilidad de desconexión (`pd_`)**: La probabilidad de que una celda pierda su línea, establecida en 0.11.

Generaremos  un  panel  de control  para  la  simulación,  inicialmente  usaremos  colab  forms

```python
# @title Automatas Celulares Chile

import numpy as np
import matplotlib.pyplot as plt
import pandas  as pd

df1 = pd.read_csv("/content/drive/MyDrive/mobile-cellular-subscriptions-by-country.csv")
df2 = pd.read_csv("/content/drive/MyDrive/Proyecto/gdp.csv")
df3 = pd.read_csv("/content/drive/MyDrive/Proyecto/urban-and-rural-population.csv")
df4 = pd.read_csv("/content/drive/MyDrive/Proyecto/co-emissions-per-capita.csv")

data =pd.read_csv("/content/drive/MyDrive/Proyecto/mobile-cellular-subscriptions-per-100-people.csv")
data_ = data[(data["Entity"]=="Chile") & (data["Year"]>=1990)]


data_.reset_index(inplace=True)

# initial  parameters

# @markdown  #  Descripción de  los  parametros
# @markdown | parametro | Descripción  | valor |
# @markdown  |------|-------------------|------------|
# @markdown  | size |     Tamaño  de la  matriz       |   250 |
# @markdown  | years |    Años en la  simulación     |   33 |
# @markdown  | pa_ |    Factor  de adquisición       |   0.61 |
# @markdown  | pd_ |    Probabilidad desconexión       |  0.11 |

population = 11110740
lines = 13921
size = 250
years = 33
pa_ = 0.61
pd_=0.11

ca = np.zeros((size,size))
pinit =lines/population

fig,ax =plt.subplots(1,2)
plot_time = np.arange(np.datetime64(24, 'Y'), np.datetime64(24, 'Y')+years)
def init_state():
  """
  Initialize  the  initial  state
  """
  global ca,pinit,size
  ca = np.zeros((size,size))
  n=int(size**2*pinit//2)
  for i,j in zip(np.random.randint(1,size-1,n),np.random.randint(1,size-1,n)):
    ca[i,j] = 1
  pass


def  evolution(year,pa=0.01,pd=0.):
  """
  Evolution  of  the  automaton
  pa:  probability  of  birth
  """
  global ca,size
  for  i  in  range(1,size-1):
    for  j  in  range(1,size-1):
      status = ca[i,j]
      total =np.sum(ca)/100.
      if  status == 0:
        nb =np.sum(ca[i-1:i+2,j-1:j+2])
        if np.random.rand() < pa*nb/9:
          ca[i,j] = 1
        else:
          if year > 8 and year <14 and np.random.rand() < pa*nb/8:
            ca[i,j]=1
      else:
        if year > 14 and np.random.rand() < pd:
          ca[i,j] = 0
  pass


def j(y,ym):
  return  np.sum(np.power(y-ym,2))/np.shape(y)[0]

if __name__ =="__main__":
  init_state()
  totalline = []
  for i in range(years):
    evolution(i,pa=pa_,pd=pd_)
    nlines = np.sum(ca)
    totalline.append(nlines/(100))
  ax[0].imshow(ca, cmap='gray')
  ax[1].plot(data_["Mobile cellular subscriptions (per 100 people)"])
  ax[1].plot(totalline)
  ax[1].legend(["Data","Simulation"])
  print(j(np.array(totalline),np.array(data_["Mobile cellular subscriptions (per 100 people)"])))
```

### **Interpretación general**
El modelo de autómatas celulares es capaz de replicar patrones generales de adopción tecnológica, como la aparición de clústeres en las etapas iniciales y un crecimiento acelerado que luego se estabiliza. Las discrepancias observadas pueden deberse a que algunos parámetros del modelo se ajustaron de forma empírica (prueba y error), lo que no garantiza una representación perfecta de la realidad.

```python
# @title Regresion Lineal
from sklearn.linear_model import LinearRegression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Cargar los datos (ajusta el nombre del archivo y la columna correspondiente)
df_co2 = pd.read_csv("/content/drive/MyDrive/Proyecto/co-emissions-per-capita.csv")
df_chile = df_co2[(df_co2["Entity"] == "Chile") & (df_co2['Year'] >= 1990) & (df_co2['Year'] <= 2023)]

# Extraer los datos relevantes (ajusta los nombres de columnas según el archivo)
X = df_chile["Year"].values
y = df_chile["Annual CO₂ emissions (per capita)"].values

# Reshape los datos para usarlos en scikit-learn
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)

# Crear el modelo de regresión lineal
modelo = LinearRegression()

# Ajustar el modelo a los datos
modelo.fit(X, y)

# Obtener los coeficientes
w_sklearn = modelo.coef_[0][0]  # Pendiente
b_sklearn = modelo.intercept_[0]  # Intercepto

# Predicciones
y_pred_sklearn = modelo.predict(X)

# Crear el gráfico
plt.figure(figsize=(6, 4))
plt.scatter(X, y, color="pink", label="Datos reales")
plt.plot(X, y_pred_sklearn, color="gray", label="Recta de regresión")
plt.xlabel("Años")
plt.ylabel("Emisiones de CO₂ per cápita")
plt.title("Regresión Lineal (scikit-learn) - Emisiones de CO₂ en Chile")
plt.legend()
plt.show()
```

## **Análisis de Regresión Lineal**
El gráfico muestra la relación entre las emisiones de CO₂ per cápita en Chile y los años, utilizando un modelo de regresión lineal ajustado con scikit-learn. A partir de los datos observados, se puede interpretar que existe una tendencia creciente en las emisiones de CO₂ per cápita a lo largo del tiempo, lo cual sugiere que las emisiones han aumentado de manera consistente desde 1990 hasta 2023.

El modelo de regresión lineal ajustado, representado por la línea gris, captura esta tendencia general, mostrando una relación aproximadamente lineal entre ambas variables. Esto indica que el tiempo (años) es un factor que contribuye directamente al incremento de las emisiones.

## **Interpretación:**
De este análisis podemos deducir que el aumento en las emisiones de CO₂ podría estar relacionado con factores como el crecimiento económico, la industrialización o el aumento en el consumo energético del país. Aunque el modelo de regresión lineal es una buena herramienta para identificar esta tendencia, es importante considerar que las emisiones también pueden estar influenciadas por otros factores no incluidos en este análisis, como políticas ambientales, cambios tecnológicos y patrones de consumo.

El modelo asume una relación lineal simple, donde:

$$y=w⋅x+b$$

donde:

* $y$ representa las emisiones de CO₂ per cápita (variable dependiente)
* $x$ representa los años (variable independiente)
* $w$ es la pendiente de la recta (cambio en las emisiones por cada año)
* $b$ es el intercepto (emisiones iniciales en el tiempo 0).


```python
# @title  ##### .

import numpy  as np
import pandas as pd

data =pd.read_csv("/content/drive/MyDrive/Proyecto/mobile-cellular-subscriptions-per-100-people.csv")
df2 = pd.read_csv("/content/drive/MyDrive/Proyecto/gdp.csv")
df3 = pd.read_csv("/content/drive/MyDrive/Proyecto/urban-and-rural-population.csv")
df4 = pd.read_csv("/content/drive/MyDrive/Proyecto/co-emissions-per-capita.csv")

dfnew =pd.merge(data,df2, on='Entity')
dfnew =dfnew[dfnew['Year_x']==dfnew['Year_y']]
dfnew =dfnew.drop(columns=['Code_y','Year_y'])
dfnew =pd.merge(dfnew,df4, on='Entity')
dfnew =dfnew[dfnew['Year_x']==dfnew['Year']]
dfnew =dfnew.drop(columns=['Code','Year'])
dfnew =pd.merge(dfnew,df3, on='Entity')
dfnew =dfnew[dfnew['Year_x']==dfnew['Year']]
dfnew =dfnew.drop(columns=['Code','Year'])
dfnew =dfnew.rename(columns={'Mobile cellular subscriptions (per 100 people)':'Mobile-Lines'})
dfnew =dfnew.rename(columns={'Code_x':'Code'})
dfnew =dfnew.rename(columns={'Year_x':'Year'})
dfnew =dfnew.rename(columns={'GDP per capita, PPP (constant 2017 international $)':'GDP'})
dfnew =dfnew.reset_index(drop=True)
dfnew.head()

# Subset dfnew to create the equivalent of df_chile
df_chile = dfnew[(dfnew["Entity"] == "Chile") & (dfnew['Year'] >= 1990) & (dfnew['Year'] <= 2023)]

# Now iterate and cast to float using the correct dataframe 'df_chile'
for chi in ['Mobile-Lines','GDP','Urban population','Rural population']:
  df_chile.loc[:,chi]=df_chile[chi].astype(float)

df_chile.head()
```
